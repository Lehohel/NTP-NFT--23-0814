```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
strk4 = pd.read_csv("./stRk4/data/output_stRK4.csv")
mtrk4 = pd.read_csv("./mtrk4/data/output_mtrk4.csv")
cuRk4 = pd.read_csv("./cuRk4/data/output_cuRK4.csv")
stABM = pd.read_csv("./stABM/data/output_stABM.csv")
mtABM = pd.read_csv("./mtABM/data/output_mtABM.csv")
cuABM = pd.read_csv("./cuABM/data/output_cuABM.csv")
```

```python
plt.figure()
plt.plot(cuRk4["Time [fs]"], np.unwrap(np.arctan2(cuRk4["Analytical imaginary"], cuRk4["Analytical real"])))
plt.plot(strk4["Time [fs]"], np.unwrap(np.arctan2(strk4["RK4 field imaginary"], strk4["RK4 field real"])))
plt.plot(mtrk4["Time [fs]"], np.unwrap(np.arctan2(mtrk4["RK4 field imaginary"], mtrk4["RK4 field real"])))
plt.plot(cuRk4["Time [fs]"], np.unwrap(np.arctan2(cuRk4["RK4 field imaginary"], cuRk4["RK4 field real"])))
plt.plot(stABM["Time [fs]"], np.unwrap(np.arctan2(stABM["RK4 field imaginary"], stABM["RK4 field real"])))
plt.plot(mtABM["Time [fs]"], np.unwrap(np.arctan2(mtABM["RK4 field imaginary"], mtABM["RK4 field real"])))
plt.plot(cuABM["Time [fs]"], np.unwrap(np.arctan2(cuABM["RK4 field imaginary"], cuABM["RK4 field real"])))
plt.ylim(-11.5,-11)
plt.xlim(-1, 1)
```
